```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr

import cartopy.crs as ccrs
```

```python
ds = xr.open_dataset(
    "/pscratch/sd/t/timothys/nested-eagle/v0/10percent/mmgt/inference/124b960f-923a-4af6-85d9-2cf0dfa763ce/2018-04-05T12.240h.nc", decode_timedelta=True)
```

```python
fig, ax = plt.subplots(
    figsize=(8,8),
    constrained_layout=True,
    subplot_kw={
        "projection": ccrs.Orthographic(
            central_latitude=20,
            central_longitude=-120,
        ),
    },
)

skip = 1
ax.scatter(
    ds.longitude[::skip],
    ds.latitude[::skip],
    c="gray",
    s=1/2,
    transform=ccrs.PlateCarree(),
)
fig.savefig("figures/1deg.15km.jpeg", bbox_inches="tight", dpi=300)
```

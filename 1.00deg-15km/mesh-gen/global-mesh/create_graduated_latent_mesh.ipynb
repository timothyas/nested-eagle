```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr

import xesmf
```

```python
ds2 = xesmf.util.grid_global(2, 2, cf=True, lon1=360)
ds3 = xesmf.util.grid_global(3, 3, cf=True, lon1=360)
ds4 = xesmf.util.grid_global(4, 4, cf=True, lon1=360)
```

```python
ds2.lat
```

```python
ds3.lat
```

```python
clat = np.concatenate(
    [
        np.arange(-89, -65, 2),
        np.arange(-65, -44, 3),
        np.arange(-44, 44, 4),
        np.arange(44, 65, 3),
        np.arange(65, 90, 2),
    ],
)
```

```python
clat
```

```python
cds = xr.Dataset()
cds["lon"] = ds4.lon
cds["lat"] = xr.DataArray(
    clat,
    coords={"clat": clat},
)
```

```python
clon = ds4.lon.values
```

```python
lon, lat = np.meshgrid(clon, clat)
```

```python
np.savez("latent-graduatedx2-4.global1degree.npz", lon=lon, lat=lat)
```

```python
cds.to_netcdf("latent-graduatedx2-4.global1degree.nc")
```

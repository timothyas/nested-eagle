```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import cartopy.crs as ccrs
```

```python
g104 = xr.open_dataset("/pscratch/sd/t/timothys/nested-eagle/1.00deg-15km/mse06h/experiments/training-steps/steps030k/inference-validation/obs-val/obs/20230201/06/gdas.20230201.t06z.prepbufr.nc")
g004 = xr.open_dataset("/pscratch/sd/t/timothys/nested-eagle/1.00deg-15km/mse06h/experiments/training-steps/steps030k/inference-validation/obs-val-global/obs/20230201/06/gdas.20230201.t06z.prepbufr.nc")
```

```python
fig, ax = plt.subplots(figsize=(10, 6), subplot_kw={"projection": ccrs.Robinson()})

for xds, color in zip([g004, g104], ["gray", "blue"]):
    ax.scatter(
        xds.hdr_lon,
        xds.hdr_lat,
        color=color,
        transform=ccrs.PlateCarree(),
        marker='.',
        label=f"{len(xds.nobs)}",
    )
ax.coastlines()
ax.set(title="Conventional Observational Coverage\n2023-02-01 06z")
fig.savefig(f"figures/obs_coverage.jpeg", dpi=300, bbox_inches="tight")
```

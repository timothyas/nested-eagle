```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import seaborn as sns
```

```python
plt.style.use("~/nice.mplstyle")
```

```python
metric = "mae"
error = {}
error = {
    name: xr.open_dataset(
        f"/pscratch/sd/t/timothys/nested-eagle/0.25deg-06km/baselines/{key}-forecasts-vs-hrrr-analysis/trim25/{metric}.lam.nc",
        decode_timedelta=True,
    )
    for name, key in zip(["HRRR", "GFS"], ["hrrr", "gfs"])
}
error["Nested-EAGLE 0.25\u00B0/06km"] = xr.open_dataset(f"/pscratch/sd/t/timothys/nested-eagle/0.25deg-06km/mse06h/experiments/1.0deg-24km-latent/1x16encoders/inference-validation/{metric}.nested-lam.nc", decode_timedelta=True)
#error["Nested-EAGLE 1.00\u00B0/15km"] = xr.open_dataset(f"/pscratch/sd/t/timothys/nested-eagle/1.00deg-15km/mse06h/experiments/regularization/wp01/inference-validation/{metric}.nested-lam.nc", decode_timedelta=True)
```

```python
error = {k: error[k] for k in ["Nested-EAGLE 0.25\u00B0/06km", "HRRR", "GFS"]}
```

```python
nice_names = {
    "sp": "Surface Pressure",
    "u10": "10m Zonal Wind",
    "v10": "10m Meridional Wind",
    "t2m": "2m Temperature",
    "sh2": "2m Specific Humidity",
    "gh": "Geopotential Height",
    "u": "Zonal Wind",
    "v": "Meridional Wind",
    "w": "Vertical Velocity",
    "t": "Temperature",
    "q": "Specific Humidity",
}
```

```python
def single_plot(ax, dsdict, varname, sel=None, **kwargs):
    for label, xds in error.items():
        linestyle = "-"
        if "Nested" in label:
            color = "C0"
            if "15km" in label:
                linestyle = ":"
        elif "HRRR" in label:
            color = "C1"
        elif "GFS" in label:
            color = "C2"
        plotme = xds[varname] if sel is None else xds[varname].sel(**sel)
        df = plotme.to_dataframe().reset_index()
        sns.lineplot(
            data=df,
            x="fhr",
            y=varname,
            ax=ax,
            label=label,
            color=color,
            linestyle=linestyle,
            **kwargs,
        )
    xticks = plotme.fhr.values
    xticks = xticks if len(xticks)<10 else np.concatenate([ [xticks[0]], xticks[4::4]])
    ax.set(
        ylabel=metric.upper() if ax.get_subplotspec().is_first_col() else "",
        xlabel="Forecast Hour" if ax.get_subplotspec().is_last_row() else "",
        title=nice_names[varname] if ax.get_subplotspec().is_first_row() else "",
        xticks=xticks,
    )
    ax.legend(frameon=False)
```

```python
def plot_surface_error(surface_vars, **kwargs):
    ncols = len(surface_vars)
    fig, axs = plt.subplots(1, ncols, figsize=(5*ncols, 4), constrained_layout=True)
    
    for varname, ax in zip(surface_vars, axs):
        single_plot(ax=ax, dsdict=error, varname=varname, **kwargs)
    return fig, axs
```

```python
def plot_level_vars(level_vars, **kwargs):
    
    levels = error["HRRR"].level.values
    ncols = len(level_vars)
    nrows = len(levels)
    fig, axs = plt.subplots(nrows, ncols, figsize=(5*ncols, 4.5*nrows), constrained_layout=True)

    if ncols == 1:
        axs = [axs]
    if nrows == 1:
        axs = [axs]
    

    sel = kwargs.pop("sel", {})
    for level, axr in zip(levels, axs):
        for varname, ax in zip(level_vars, axr):
            
            sel["level"] = level
            single_plot(ax=ax, dsdict=error, varname=varname, sel=sel, **kwargs)
            ax.legend(title=f"{level} hPa", frameon=False)
    return fig, axs
```

```python
fig, axs = plot_surface_error(["t2m", "u10"], sel={"fhr": slice(49)}, marker="o")
fig.savefig(f"figures/t2m_u10_{metric}_vs_hrrr_analysis.jpeg", dpi=300, bbox_inches="tight")
```

```python
fig, axs = plot_surface_error(["sp", "u10", "v10", "t2m", "sh2"], sel={"fhr": slice(49)}, marker="o")
fig.savefig(f"figures/surface_{metric}_vs_hrrr_analysis_48h.jpeg", dpi=300, bbox_inches="tight")
```
